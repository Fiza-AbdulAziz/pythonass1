```python
def to_uppercase_manual(s):
    result = ""
    for char in s:
        if 'a' <= char <= 'z':
            # Convert lowercase to uppercase using ASCII difference
            result += chr(ord(char) - 32)
        else:
            result += char
    return result

# Example usage
input_str = "hello world"
print("Uppercase (manual):", to_uppercase_manual(input_str))  # Output: "HELLO WORLD"
```

```python
#5. Count the number of vowels in a string.
def count__vowels(s):
    vowels ='aeiouAEIOU'
    count =0
    for char in s:
        if char in vowels:
            count +=1
    return count
input_str = "hElLO wOrLD"
print("count vowels",count__vowels(input_str))

def count_vowels_lc(s):
    return sum(1 for char in s if char.lower() in "aeiou")

# Example usage
input_str = "Hello World"
print("Number of vowels (LC):", count_vowels_lc(input_str))  
```
